```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
sns.set()
```

```python
data_movies_final = pd.read_csv('.\Datasets\movies_dataset_limpio.csv')

```

```python
df_movies_final = pd.DataFrame(data_movies_final)
```

1- Voy a desanidar la columna genres para quedarme solo con los nombres de los géneros de cada película

```python
# Imprimo para ver qué trae la columna genres
print(df_movies_final['genres'].head())
```

```python
# Itero sobre cada fila para extraer solo el primer genero y agregar una columna con él

for index, row in df_movies_final.iterrows():
    genre = json.loads(row['genres'].replace("'", "\""))
    if genre:
        df_movies_final.at[index, 'genre'] = genre[0]['name']
    else:
        df_movies_final.at[index, 'genre'] = None

```

```python
# Elimino la columna original "genres"
df_movies_final = df_movies_final.drop(columns=["genres"])
```

2- Ahora voy a desanidar la columna "production_countries" para quedarme solo con el país

```python
# Iterar sobre cada fila del DataFrame

for index, row in df_movies_final.iterrows():
    country = json.loads(row['production_countries'].replace("'", "\""))
    if country:
        df_movies_final.at[index, 'country'] = country[0]['name']
    else:
        df_movies_final.at[index, 'country'] = None
```

```python
# Elimino la columna original "production_countries"
df_movies_final = df_movies_final.drop(columns=["production_countries"])
```

4- Reducido el dataset de Credits, procedo a su limpieza

```python
df_credits_filtered.head()
```

```python
# Extraigo el nombre del primer actor que aparece en la lista de diccionarios de la columna cast
# Por algún motivo que no logré ubicar solo reconoce al actor principal en unas 100 de las 900 películas

# Función para extraer el nombre del primer actor
def get_first_actor(cast_str):
    try:
        cast_list = json.loads(cast_str.replace("'", "\""))
        if cast_list:
            return cast_list[0]['name']
        return None
    except json.JSONDecodeError:
        return None

# Aplicar la función a cada fila de la columna 'cast' y crear una nueva columna 'principal_cast'
df_credits_filtered['principal_cast'] = df_credits_filtered['cast'].apply(get_first_actor)

```

```python
# Extraigo el nombre del director de la columna crew

# Función para extraer el nombre del director y el job
def get_director_info(crew_str):
    try:
        crew_list = json.loads(crew_str.replace("'", "\""))
        for member in crew_list:
            if member['job'] == 'Director':
                return pd.Series([member['job'], member['name']])
        return pd.Series([None, None])
    except json.JSONDecodeError:
        return pd.Series([None, None])

# Aplico la función a cada fila de la columna 'crew' y creo nuevas columnas 'job' y 'director'
df_credits_filtered[['job', 'director']] = df_credits_filtered['crew'].apply(get_director_info)
```

```python
# Voy  a eliminar las columnas que no necesito
df_credits_filtered = df_credits_filtered.drop(columns=["cast", "crew", "job"])
```

```python
# Concateno los dataframes por la columna 'id'
df_movies_final = pd.merge(df_movies_final, df_credits_filtered, on='id', how='inner')

```
